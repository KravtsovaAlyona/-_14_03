```python
df=pd.read_csv('person.csv.csv')
df
```

```python
df['registration_date'] = pd.to_datetime(df['registration_date'])
df['created_at'] = pd.to_datetime(df['created_at'])
df['updated_at'] = pd.to_datetime(df['updated_at'])
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Гистограмма распределения возраста')
plt.xlabel('Возраст')
plt.ylabel('Количество людей')
plt.grid(axis='y', alpha=0.75)
plt.xticks(range(0, df['age'].max() + 1, 5))
plt.show()
```

```python
df1=df
# Извлекаем название города (первая часть 'address', разделенная запятой)
df1['city'] = df['address'].apply(lambda x: x.split(',')[0])

# Считаем количество проживающих в каждом городе
city_counts = df1['city'].value_counts()


top_cities = city_counts.head(5)


plt.figure(figsize=(9, 5))
top_cities.plot(kind='bar', color='skyblue', edgecolor='black')

plt.xlabel("Город")
plt.ylabel("Количество проживающих")
plt.title("Топ-5 городов по количеству проживающих")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.show()
```

```python
df2=df
# Группируем по месяцам и считаем количество регистраций
registrations_per_month = df2.groupby(df2['registration_date'].dt.to_period('M')).size()

plt.figure(figsize=(10, 5))
registrations_per_month.plot(kind='line', marker='o', linestyle='-', color='b', label="Регистрации")

plt.xlabel("Месяц")
plt.ylabel("Количество регистраций")
plt.title("Количество регистраций по месяцам за последний год")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

plt.show()
```
