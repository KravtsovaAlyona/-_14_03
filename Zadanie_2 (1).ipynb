```python
df=pd.read_csv('person_count_by_city.csv')
df
```

```python
top_cities = df.sort_values(by='name', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(top_cities['city'], top_cities['name'], color='skyblue')
plt.title('Топ-10 городов по количеству проживающих')
plt.xlabel('Города')
plt.ylabel('Население')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['name'], bins=5, color='lightcoral', edgecolor='black')
plt.title('Гистограмма распределения населения по городам')
plt.xlabel('Население')
plt.ylabel('Количество городов')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
min_population = df['name'].min()
max_population = df['name'].max()
line_data = pd.DataFrame({
    'Города': ['Минимум', 'Максимум'],
    'Население': [min_population, max_population]
})
plt.figure(figsize=(10, 6))
plt.plot(line_data['Города'], line_data['Население'], marker='o', color='purple', linestyle='--')
plt.title('Минимальное и максимальное количество людей в городах')
plt.xlabel('Тип')
plt.ylabel('Население')
plt.grid()
plt.show()
```
